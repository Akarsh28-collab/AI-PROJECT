```python
# ================================
# 1. IMPORT LIBRARIES
# ================================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# ================================
# 2. LOAD DATA
# ================================
df = pd.read_csv('Titanic-Dataset.csv')

# ================================
# 3. BASIC INFO
# ================================
print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# ================================
# 4. MISSING VALUES
# ================================
print("\nMissing Values:")
print(df.isnull().sum())

# ================================
# 5. UNIVARIATE ANALYSIS
# ================================

# ---------- Survived ----------
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()

print("\nObservation (Survived):")
print("- Majority passengers did NOT survive")
print("- Dataset is slightly imbalanced")


# ---------- Sex ----------
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()

print("\nObservation (Sex):")
print("- More males than females onboard")


# ---------- Pclass ----------
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()

print("\nObservation (Pclass):")
print("- Most passengers belong to 3rd class")


# ---------- Age ----------
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

print("\nObservation (Age):")
print("- Most passengers are between 20–40 years")
print("- Distribution is slightly right-skewed")


# ---------- Age Boxplot ----------
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title("Age Boxplot")
plt.show()

print("\nObservation (Age Boxplot):")
print("- Presence of outliers (elderly passengers)")
print("- Median age around late 20s")


# ---------- Fare ----------
plt.figure(figsize=(6,4))
sns.histplot(df['Fare'], kde=True)
plt.title("Fare Distribution")
plt.show()

print("\nObservation (Fare):")
print("- Highly right-skewed distribution")
print("- Few passengers paid very high fares")


# ---------- Fare Boxplot ----------
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()

print("\nObservation (Fare Boxplot):")
print("- Many outliers present")
print("- Majority fares are low")


# ---------- Embarked ----------
plt.figure(figsize=(6,4))
sns.countplot(x='Embarked', data=df)
plt.title("Embarked Distribution")
plt.show()

print("\nObservation (Embarked):")
print("- Most passengers boarded from Southampton (S)")


# ================================
# 6. FINAL INSIGHTS
# ================================
print("\nFINAL INSIGHTS:")
print("- Majority passengers did not survive")
print("- More males than females onboard")
print("- Most passengers were in 3rd class")
print("- Age mostly between 20–40 with some outliers")
print("- Fare distribution is highly skewed")
print("- Missing values present in Age, Cabin, Embarked")
```
