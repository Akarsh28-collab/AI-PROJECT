```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd

all_quotes=[]
all_name=[]
all_tags=[]

for i in range(1,11):
    url=f"https://quotes.toscrape.com/page/{i}/"
    response=requests.get(url)
    soup=BeautifulSoup(response.text,"html.parser")
    quot=soup.find_all("div",class_="quote")

    for q in quot:
        quote=q.find("span",class_="text").text.strip()
        name=q.find("small",class_="author").text.strip()
        tags = [tag.text.strip() for tag in q.find_all("a", class_="tag")]
        tags = ", ".join(tags)
        all_quotes.append(quote)
        all_name.append(name)
        all_tags.append(tags)
        
df = pd.DataFrame({
    "QUOTE": all_quotes,
    "AUTHOR": all_name,
    "TAGS": all_tags
})
df.to_csv("quotes_data.csv", index=False) 
```
